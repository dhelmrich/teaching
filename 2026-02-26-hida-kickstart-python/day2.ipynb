## List Comprehension

```python
xdata = range(10)
ydata = [x**2 for x in xdata]
plt.plot(xdata, ydata)
```

```python
ydata = [x**2 for x in range(10)]
plt.plot(range(10), ydata)
```

```python
samples = ["sampleA", "sampleB", "sampleC", "sampleA_QualityFlags", "sampleB_QualityFlags", "sampleC_QualityFlags"]
samples = [sample for sample in samples if not "QualityFlags" in sample]
# this list comprehension is equivalent to:
result_samples = []
for sample in samples :
    if not "QualityFlags" in sample :
        result_samples.append(sample)
print(samples)
print(result_samples)
```
