# Matplotlib Introduction

```python
from matplotlib import pyplot as plt
import datetime
from random import random
```

```python
air_temperature = [
5.0, 5.6, 5.6, 5.6, 5.6, 6.1, 6.7, 7.2,
7.8, 8.3, 8.3, 8.9, 7.8, 7.2, 7.2, 7.2,
6.7, 6.7, 6.7, 6.7, 6.1, 5.6, 5.0, 5.0
]
len(air_temperature)
```

```python
plt.plot(air_temperature)
plt.scatter(4, 4)
```

```python
start_time = datetime.datetime(year = 2020, month = 12, day = 31, hour = 0)
hours = [start_time + datetime.timedelta(hours=offset) for offset in range(len(air_temperature))]
hours
```

```python
range(len(air_temperature))
```

```python
plt.plot(hours, air_temperature)
```

```python
plt.title("Central Park Temperature")
plt.plot(hours, air_temperature)
plt.xlabel("Local Time")
plt.grid()
plt.ylabel("Air Temperature [°C]")
```

```python
displayed_hours = hours[::2]
x_axis_tick_labels = [hour.strftime("%H:%M") for hour in displayed_hours]
x_axis_tick_labels
```

```python
plt.title("Central Park Temperature")
plt.plot(hours, air_temperature)
plt.xlabel("Local Time")
plt.grid()
plt.ylabel("Air Temperature [°C]")
plt.xticks(ticks = displayed_hours, labels = x_axis_tick_labels,
           rotation = 45,
           horizontalalignment = "right"
          )
```

```python
average_temperature = sum(air_temperature) / len(air_temperature)
average_temperature
# range(15) # gives your 0 .. 14
day_start = 7
day_end = 17
daytime = hours[day_start:day_end + 1] # full would be hours[0:len(hours)]
daytime_temperature = air_temperature[day_start:day_end + 1]
```

```python
plt.title("Central Park Temperature")
plt.xlabel("Local Time")
plt.ylabel("Air Temperature [°C]")
plt.xticks(ticks = displayed_hours, labels = x_axis_tick_labels,
           rotation = 45,
           horizontalalignment = "right"
          )
# horizontal line, only depends on the y value
plt.axhline(y = average_temperature, linestyle="dashed", color="gray")

# shaded plot area --> The second one will "overwrite" the first one
# but axhline beats both
plt.fill_between(
    x = hours,
    y1 = air_temperature,
    y2 = min(air_temperature),
    color = "lightblue"
)
plt.fill_between(
    x = daytime,
    y1 = daytime_temperature,
    y2 = min(air_temperature),
    color = "lightyellow"
)
plt.plot(hours, air_temperature)

plt.show()
```

# Excercise 1

```python
# Initial parameters
value_count = 365                 # How many days to predict
start_day = datetime.datetime(2020, 1, 1)  # Start day of the prediction
price = 50 * (1 + random())       # Initial share price

# Prediction data
share_prices = [
price := price - 0.05 + random() / 10
for _ in range(value_count)
]

# Dates assiciated with the data
days = [
start_day + datetime.timedelta(days=current_value)
for current_value in range(value_count)
]
# change the size of the plot
plt.figure(figsize = (8,4))
plt.plot(days, share_prices)
plt.show()
```

## Step 2: Text Info

```python
# change the size of the plot
plt.figure(figsize = (8,4))
plt.plot(days, share_prices)
plt.title("Profit Inc. Share Prices (predicted)")
plt.xlabel("Trade day")
plt.ylabel("Share Price [€]")
plt.show()
```

## Step 3: Average Share Price

```python
# average share price
avg_price = sum(share_prices) / len(share_prices)

# change the size of the plot
plt.figure(figsize = (8,4))

# data plotting
plt.plot(days, share_prices)
plt.axhline(avg_price, color="gray", linestyle="dashed")

# information and plt.show
plt.title("Profit Inc. Share Prices (predicted)")
plt.xlabel("Trade day")
plt.ylabel("Share Price [€]")
plt.show()
```

## Step 4: Price Range

```python
# Additional information: min and max prices and their positions
min_price = min(share_prices)
max_price = max(share_prices)
min_idx = share_prices.index(min_price)
max_idx = share_prices.index(max_price)
min_day = days[min_idx]
max_day = days[max_idx]
print("Max/Min Prices are:", min_price, max_price)
print("At position:", min_idx, max_idx)
print("... in x-axis units:", min_day, max_day)

# Additional information: average share price
avg_price = sum(share_prices) / len(share_prices)

# change the size of the plot
plt.figure(figsize = (8,4))

# data plotting
plt.plot(days, share_prices)
plt.axhline(avg_price, color="gray", linestyle="dashed")

# use the vlines method (like in the documentation)
plt.vlines(min_day, avg_price, min_price, color="red")
plt.vlines(max_day, avg_price, max_price, color="green")

# Shaded area
leftmost_idx = min(min_idx, max_idx)
rightmost_idx = max(min_idx, max_idx)
plt.fill_between(
    days[leftmost_idx:rightmost_idx+1],
    avg_price,
    share_prices[leftmost_idx:rightmost_idx+1],
    color="lightgray")

# information and plt.show
plt.title("Profit Inc. Share Prices (predicted)")
plt.xlabel("Trade day")
plt.ylabel("Share Price [€]")
print(plt.gcf())
plt.show()
```
